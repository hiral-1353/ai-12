# Lists

```python
# Create a list of 10 numbers
list = [-5, -4, -3, -2, -1, 1, 2, 3, 4, 5]

# Print only even numbers
new_list = [i for i in list if i % 2 == 0]

print(new_list)

# Replace all negative numbers with 0
zero_list = [i if i >= 0 else 0 for i in list]
print(zero_list)
```

```python
# Input a list of numbers from the user
list = input("Enter a list of numbers, separated by commas (no space): ").split(",")
print(list)

# Print the second largest number
list.sort()
print(list)

print("The second largest number is:", list[-2])

# Count how many numbers are greater than 50
```
