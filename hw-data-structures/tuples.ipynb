# Tuples

```python
# Create a tuple of 6 elements
tup = (1, 2, 3, 4, 5, 6)
print("Tuple:", tup)

# Convert it into a list
tup_list = list(tup)
print("List:", tup_list)

tup_list.insert(6, 7)
tup_list.insert(7, 8)
print("Updated List:", tup_list)

# List to tuple
new_tup = tuple(tup_list)
print("New Tuple:", new_tup)
```

```python
tup = (10, 20, 30, 20, 40)

# Count occurences of 20
count_20 = tup.count(20)
print("Count of 20:", count_20)

# Maximum Value
max_value = max(tup)
print("Maximum Value:", max_value)
```
