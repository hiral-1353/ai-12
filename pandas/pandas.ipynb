```python
# Empty Series
import pandas as pd
import numpy as np

sr = pd.Series()

print(sr)
```

```python
# Series using Numpy array
import pandas as pd
import numpy as np

data = np.array(["a", "b", "c", "d"])
sr = pd.Series(data)

print(sr)
```

```python
# Series using list
import pandas as pd

data = ["a", "b", "c", "d"]
sr = pd.Series(data)

print(sr)
```

```python
# Series using scalar value
import pandas as pd

sr = pd.Series(24, index=[0, 1, 2, 3])

print(sr)
```

```python
# Empty dataframe
import pandas as pd
df = pd.DataFrame()
print(df)
```

```python
# Dataframe using list
import pandas as pd
data = ["a", "b", "c", "d"]
df = pd.DataFrame(data)
print(df)
```

```python
# Dataframe using Numpy arrays
import pandas as pd
import numpy as np

data = np.array(["a", "b", "c", "d"])
data1 = np.array([1, 2, 3, 4])
df = pd.DataFrame({"col1": data, "col2": data1})

print(df)
```

```python
# Dataframe from Numpy arrays using list
import pandas as pd
import numpy as np

data = np.array(["a", "b", "c", "d"])
data1 = np.array([1, 2, 3, 4])

df = pd.DataFrame([data, data1], columns=["A", "B", "C", "D"])

print(df)
```

```python
# Dataframe using dictionary
import pandas as pd

data = {
    "Books": ["Mistborn", "The Hobbit", "Resurrection Walk"],
    "Author": ["Brandon Sanderson", "J.R.R. Tolkien", "Michael Connelly"],
    "Year": [2006, 1937, 2023]
}

df = pd.DataFrame(data)

print(df)
```
