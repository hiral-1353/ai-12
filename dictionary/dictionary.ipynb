```python
student = {
    "name": "John Doe",
    "class": 11,
    "marks": 95
    }
print(student)
```

```python
# Dictionary Creation

# Method 1
d1 = {"name": "Alice", "age": 16, "city": "New York"}
print(d1)

# Method 2
d2 = dict(name="Bob", age=17, city="Los Angeles")
print(d2)

# Empty Dictionary
d3 = {}
print(d3)
```

```python
student = {
    "name": "John Doe",
    "class": 11,
    "marks": 95
    }

# Access using Key
print(student["name"])

# Access using get() method
print(student.get("class"))
```

```python
student = {
    "name": "John Doe",
    "class": 11,
    "marks": 95
    }

# Adding a new key
student["age"] = 17
print(student)

# Updating an existing key
student["marks"] = 98
print(student)
```
