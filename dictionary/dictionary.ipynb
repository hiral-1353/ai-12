```python
student = {
    "name": "John Doe",
    "class": 11,
    "marks": 95,
    "age": 17
    }

# Removing elements
student.pop("class")
print("Removing class: \n",student)
    
# Pop
student.popitem() # Removing last key
print("Removing last key: \n",student)

# Delete
del student["marks"]
print("Removing marks: \n",student)
```

```python
# Functions
d = {
    "a": 1,
    "b": 2,
}

print(d.keys())
print(d.values())
print(d.items())
```

```python
# Loop in Dictionary
d = {
    "a": 1,
    "b": 2,
}

print("Items: ")
for i in d:
    print(i)

print("Keys: ")
for i in d.keys():
    print(i)

print("Values: ")
for i in d.values():
    print(i)

print("Items: ")
for i in d.items():
    print(i)
```

```python
# Nested Dictionary
students = {
    "s1" : {"name": "Anakin Skywalker", "age": 19, "marks": 95},
    "s2" : {"name": "Luke Skywalker", "age": 18, "marks": 98},
}

print(students)

print(students["s1"]["name"])

# Shortcut for nested dictionary
d = {x : x*x for x in range(5)}
print(d)

# Membership Operator
d1 = {
    "a":1,
    "b":2,
}
print("a" in d1)
print("c" not in d1)
```
